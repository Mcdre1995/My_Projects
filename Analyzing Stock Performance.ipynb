```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

## Using the yfinance Library to Extract Stock Data

Using the `Ticker` module we can create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, here the company is Apple and the ticker symbol is `AAPL`.

```python
# Specify the stock symbol
stock_symbols = "AAPL TSLA AMZN AMD"

# Create a Ticker object
stock_ticker = yf.Tickers(stock_symbols)
```

```python
# Get stocks information
apple_info=stock_ticker.tickers["AAPL"].info
print(apple_info, "\n")  #info for Apple Stocks

tesla_info=stock_ticker.tickers["TSLA"].info
print(tesla_info, "\n") #info for Tesla Stocks

amazon_info=stock_ticker.tickers["AMZN"].info
print(amazon_info, "\n")   #info for Amazon Stocks

amd_info= stock_ticker.tickers["AMD"].info
print(amd_info)     #info for AMD Stocks
```

```python
# Fetch historical data for Apple stock
apple_historical_data = stock_ticker.tickers["AAPL"].history(period="1y")  # Fetch data for the past 1 year

# Display the fetched data
print(apple_historical_data.head())
```

```python
#Fetching Historical Data for TEsla Stock
tesla_historical_data= stock_ticker.tickers["TSLA"].history(period="1y") #fetch data for the past 1 year

#display the fetched data
print (tesla_historical_data)
```

```python
#Fetching Historical Data for TEsla Stock
amazon_historical_data= stock_ticker.tickers["AMZN"].history(period="1y") #fetch data for the past 1 year

#display the fetched data
print (amazon_historical_data)
```

```python
#Fetching Historical Data for TEsla Stock
amd_historical_data= stock_ticker.tickers["AMD"].history(period="1y") #fetch data for the past 1 year

#display the fetched data
print (amd_historical_data)
```

```python
# Plotting stock prices on the same axis 

#creating a figure and axis
fig,ax = plt.subplots()

#Plotting each column on the same axis
ax.plot(apple_historical_data['Close'], label='AAPL')
ax.plot(tesla_historical_data['Close'], label='TSLA ')
ax.plot(amazon_historical_data['Close'], label='AMZN ')
ax.plot(amd_historical_data['Close'], label='AMD ')

#add legend and labels
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Price")
ax.set_title("Watchlist")

#removing the spines 
ax.spines["top"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["bottom"].set_visible(False)

#show the plot
plt.show()
```

```python
# Plotting stock prices on different subplots

#creating a figure and subplots
fig,ax = plt.subplots(1,4,figsize=(15,4))

#Plotting each column on the same axis
ax[0].plot(apple_historical_data['Close'], label='AAPL')
ax[1].plot(tesla_historical_data['Close'], label='TSLA')
ax[2].plot(amazon_historical_data['Close'], label='AMZN')
ax[3].plot(amd_historical_data['Close'], label='AMD')

#add legend and labels
for a in ax:
   
    a.legend(loc="upper left")
    a.set_xlabel("Date")
    a.set_ylabel("Price")

#adjust layout
plt.tight_layout()

#show the plot
plt.show()
```

### Extracting Dividends

Dividends are the distribution of a companys profits to shareholders. In this case they are defined as an amount of money returned per share an investor owns. Using the variable `dividends` we can get a dataframe of the data. The period of the data is given by the period defined in the 'history` function.

```python
#getting Apple dividend
apple_div=stock_ticker.tickers["AAPL"].dividends
apple_div, apple_div.shape
```

```python
#getting Tesla dividend
tesla_div=stock_ticker.tickers["TSLA"].dividends
tesla_div, tesla_div.shape
```

```python
#getting Amazon dividend
Amazon_div=stock_ticker.tickers["AMZN"].dividends
Amazon_div, Amazon_div.shape
```

```python
#getting Amd dividend
AMD_div=stock_ticker.tickers["AMD"].dividends
AMD_div, AMD_div.shape
```

```python
#plotting the Dividend Data in one axis

fig, ax= plt.subplots()

#plotting each dividend
ax.plot(apple_div,label="AAPL")
ax.plot(tesla_div, label="TSLA")
ax.plot(Amazon_div, label="AMZN")
ax.plot(AMD_div, label="AMD")

#add legends and Labels
ax.legend()
ax.set_xlabel("Date")
ax.set_ylabel("Dividends")

#show the plot 
plt.show
```

```python
# Calculate daily returns
apple_historical_data['Daily_Return'] = apple_historical_data['Close'].pct_change()

# Display the data with daily returns
print(apple_historical_data.head())
```
